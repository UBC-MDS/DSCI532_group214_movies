**Summary for Jumbo**

If you are nascent to investing in the movie industry, your search ends right here! As a new investor, there are 4 fundamental questions you must know to estimate what your money is worth.

	1. What genre has been most profitable?
	2. How much does audience reception affect profit?
	3. What are the most and least popular genres that producers typically engage in?

Welcome to our Interactive Movie Dashboard! It is quite simple and involves understanding 2 levers which will make your exploration fun and insightful.

	1. Use the slider to observe the evolution of the industry over the select time period.
	2. Use the drop down menu to look up the top XX genres


**Effectiveness:**

**What elements of the dashboard were effective?**

1. The plot helps to answer the three core questions. The user can succesfully answer the research questions with the dashboard.

2. The plots are interactive and change based on the conditions (genre, year).

3. It is well laid out with a header, body and footer. 

4. Provides a summary of impactful movies over the filtered conditions.

5. The colors in the graph are aesthetically pleasing. 

6. We control the options that the users could experiment with so that the user doesn't break the dashboard.

7. We didn't call it a 'dashboard'!

**How could have we made it even better?**

1. The summary statistics section could have been made with a border/box to delineate the section from the rest and also aesthetically mapped to look similar to the rest of the dashboard.

2. We would have liked to have all 4 sections align to look square.

3. We have shown only the major genre and not the marriage between the creative type with it.

4. We haven't delineated the gross obtained from US and outside of the US. Rather, the WW gross has been used. So, for someone targetting just the US audience might want to look at the difference between the two.

**Summary of Feedback**

1. The general layout and functionality of the app was well received. No one had any particularly serious issue with how the app in general functioned or was presented.

2. Some thought the introduction section was too long or text heavy. Some also thought the research question section in the introduction could be rewritted to be more general and inviting.

3. The first graph could be slightly reworked to be a dodged bar chart or a faceted bar chart to make the comparison between domestic and international box office more clear. There were also suggestions to implement a local filter to see only domestic or only international box office if they so choose. Some thought the graph was too US centric and the option to include gross from other countries/continents would be nice.

4. There were suggestions for the second graph to be interactive, and allow filtering based on genre to be done on all the graphs by selecting a bar.

5. The heatmap is too scarce to be used on less popular genres and on smaller year ranges. Suggestion to make the bin width smaller so the data appear to be more dense, or to scrap the heatmap entirely for possibly a boxplot or line chart.

6. Mixed opinions on the dynamic text. Some thought it was very unique and informative but others thought it was too text heavy and didn't look dynamic at all, and aesthetically didn't fit with the rest of the app.

7. Suggestions to put the filters on the side or at the bottom as well so they're always in sight regardless of how the page is scrolled.

**Reflection on Feedback**

Based on how people dove into using our app, we would say that functionality wise our app is very intuitive, with 2 simple filters that globally change the charts/text. However some thought that the it was unclear that the summary text was dynamic. People used the app in the way we expected, overall.

Some common opinions include the scarcity of the heatmap under certain selection criteria, the desire for the filtering of domestic/international box offices in the boxplot, and just making the dynamic text fit in with the rest of the app better. People also desired interactivity of the graphs.

Changing the graph type or graph parameters slightly is reasonable and doable in a short timeframe. Adding another filter option for one graph is also reasonable, although it may affect the layout more than anticipated. Adding interactivity would be somewhat more difficult and time consuming. Requests like the inclusion of box office figures from other countries specifically or continents is not possible with our current dataset.

Although we were generally aware of the strengths and weaknesses of our app in its current state, the feedback we received was overall quite useful and helped us create priorities for what changes we want to make. The fly on the wall portion for more complex apps could be very useful to see what people find intuitive and what isn't.

Overall we believe that the feedback activity helped lead to an improved and more user-friendly app.
